```python
import logging

import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colormaps

logging.getLogger("matplotlib.font_manager").setLevel(logging.ERROR)

plt.rcParams["figure.figsize"] = (10, 5)
rng = np.random.default_rng(20260530)

hand_drawn_arrow = {
    "arrowstyle": "->",
    "connectionstyle": "arc3,rad=0.15",
    "shrinkA": 0,
    "shrinkB": 4,
}
```

```python
x = np.linspace(0, 10, 100)

with plt.xkcd():
    _, ax = plt.subplots()
    ax.plot(x, np.sin(x))
    ax.set_title("Whoo Hoo!!!")
    plt.show()
```

```python
def sin_or_cos(values):
    return np.where(values > 5, np.sin(values), np.cos(values))


x = np.linspace(0, 10, 200)
y = sin_or_cos(x)

with plt.xkcd():
    _, ax = plt.subplots()
    ax.plot(x, y)
    ax.annotate(
        "WHERE WE PREDICTED\nTHE MARKET WOULD GO",
        xy=(5.1, np.cos(5)),
        xytext=(2.3, 0.55),
        arrowprops=hand_drawn_arrow,
    )
    ax.annotate(
        "WHERE THE MARKET\nACTUALLY WENT",
        xy=(5.2, np.sin(5) + 0.05),
        xytext=(6.2, -0.55),
        arrowprops=hand_drawn_arrow,
    )
    ax.set_title("Forecasting")
    plt.show()
```

```python
sample_size = 50
x = rng.random(sample_size)
y = rng.random(sample_size)
colors = colormaps["Set3"](np.linspace(0, 1, sample_size))
area = np.pi * (7 * rng.random(sample_size)) ** 3

with plt.xkcd():
    _, ax = plt.subplots()
    ax.scatter(x, y, s=area, alpha=0.7, color=colors, edgecolor="black", linewidth=2)
    ax.set(title="Bubbles", xlabel="Noise", ylabel="Chaos")
    ax.annotate(
        "ME",
        xy=(x[0], y[0]),
        xytext=(x[0] + 0.05, y[0] - 0.05),
        arrowprops=hand_drawn_arrow,
    )
    ax.annotate(
        "MY FAVORITE BUBBLE",
        xy=(x[1], y[1]),
        xytext=(x[1] + 0.05, y[1] - 0.05),
        arrowprops=hand_drawn_arrow,
    )
    plt.show()
```

```python
bar_count = 20
theta = np.linspace(0.0, 2 * np.pi, bar_count, endpoint=False)
radii = 10 * rng.random(bar_count)
width = 2 * rng.random(bar_count)

with plt.xkcd():
    _, ax = plt.subplots()
    bars = ax.bar(theta, radii, width=width, bottom=0.0)

    for radius, bar in zip(radii, bars):
        bar.set_facecolor(colormaps["viridis"](radius / 10))
        bar.set_alpha(0.6)

    ax.set(title="Not Quite Polar", xlabel="Angle-ish", ylabel="Radius-ish")
    plt.show()
```

```python
bar_count = 20
theta = np.linspace(0.0, 2 * np.pi, bar_count, endpoint=False)
radii = 10 * rng.random(bar_count)
width = np.pi / 2 * rng.random(bar_count)

with plt.xkcd():
    _, ax = plt.subplots(subplot_kw={"projection": "polar"})
    bars = ax.bar(theta, radii, width=width, bottom=0.0)

    for radius, bar in zip(radii, bars):
        bar.set_facecolor(colormaps["plasma"](radius / 10))
        bar.set_alpha(0.6)

    ax.set_title("Actually Polar")
    plt.show()
```
