### Rabbits

Exponential growth can turn a simple population model into an absurdly large number very quickly. This notebook explores rabbit population growth under an aggressive monthly rate, compares the results to planetary-scale quantities, and visualizes how sensitive the claim is to assumptions.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (8, 6)
```

**Exponential Leporidae**

Rabbits can have three or four litters annually, Animal Diversity Web reports. Each litter can contain between one and 12 babies, with the average being five. 

4 * 5 = 20

(2 + 20)/2 = 11

11^(1/12) = 1.22

We need 1.3 for the below to work out:

> Left unchecked, 1,000 rabbits in 5 years, become 7-billion, the human population of the World. - Neil deGrasse Tyson

> After 15 years, a "land-ocean" of rabbits fills to one-kilometer depth across all of Earth’s continents.

```python
# data
initial_population = 2  # Starting with one breeding pair
growth_rate = 1.3  # Assumed growth rate per month
months = np.arange(0, 12*18, 1)
population = initial_population * (growth_rate ** months)

# First Year
plt.plot(months[:13], population[:13], label='Rabbit Population')
plt.xlabel('Months')
plt.ylabel('Rabbit Population')
plt.title('Rabbit Population - First Year')
plt.ylim(0, 40)
month_labels = [f"{month}" for month in months]
plt.xticks(ticks=months[:13], labels=month_labels[:13]);
plt.show()
# plt.yscale('log')


# Plotting
plt.plot(months, population, label='Rabbit Population')
plt.xlabel('Years')
plt.ylabel('Rabbit Population')
plt.title('Years Until Rabbit\'s Exceed the Mass of the Earth')
# plt.yscale('log')

# Mass of the Earth for comparison
earth_mass = 5.972e+24 # kg
rabbit_mass = 1.5 # kg
earth_mass_in_rabbits = 4e24
plt.axhline(earth_mass_in_rabbits, linestyle='dashed', lw=1, color='black', label="Earth Mass (in Rabbits)")

plt.ylim(0, None)

# Convert month ticks to year ticks
year_ticks = np.arange(0, 12*18 + 1, 12)
year_labels = [f"{int(month / 12)}" for month in year_ticks]
plt.xticks(ticks=year_ticks, labels=year_labels);
plt.legend();
```
