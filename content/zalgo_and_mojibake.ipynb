```python
def mojibake(text: str) -> str:
    """Intentionally corrupt UTF-8 text into common mojibake."""
    return text.encode("utf-8").decode("latin-1")

naive = mojibake("naïve")
naive
```

```python
def dezalgofy(text: str) -> str:
    """Remove the combining Unicode marks used to make Zalgo text."""
    cleaned = "".join(
        character
        for character in text
        if not unicodedata.category(character).startswith("M")
    )
    return unicodedata.normalize("NFC", cleaned)


message = "Your Name Here!"
zalgofied = z.zalgofy(message)
print(zalgofied)

recovered = dezalgofy(zalgofied)
print(recovered)

assert message == recovered
```
