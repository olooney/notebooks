### Heegner Numbers

[Heegner numbers](https://en.wikipedia.org/wiki/Heegner_number) are real numbers with simple expressions that are suprisingly close to integers. So close, in fact, that Python has trouble even representing the difference. This notebook shows that Python and NumPy lack the necessary precision to represent one particularly famous Heegner number, Ramanujan's Constant $e^{\pi\sqrt{163}}$, but the [mpmath](https://mpmath.org/) library is able to handle it with ease.



```python
# Native Python can't represent the number precisely enough:
from math import exp, pi, sqrt

exact = exp(pi * sqrt(163))
rounded = round(exact)

print(exact)
print(rounded)
print(rounded - exact)
```

```python
# Numpy can't handle it either:
from numpy import exp, pi, sqrt

exact = exp(pi * sqrt(163))
rounded = round(exact)

print(exact)
print(rounded)
print(rounded - exact)
```

```python
# But MPMath (Multiple-Precision Math) is, if we increase precision:
import mpmath
from mpmath import exp, pi, sqrt, nint as round

# increase number of  decimal places
mpmath.mp.dps = 50

exact = exp(pi * sqrt(163))
rounded = round(exact)

print(exact)
print(rounded)
print(rounded - exact)
```
