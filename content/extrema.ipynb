```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(25, 525, 25)
```

```python
y = (x**2)*(750-x)
y = 0.2 * y / np.max(y)
```

```python
y2 = y/x
y2 = y2 / np.mean(y2)
```

```python
plt.plot(x, y, label='Response Rate')
plt.xlabel("Incentive")
plt.ylabel("Response Rate")
plt.title("Response Rate vs. Incentive")
```

```python
plt.plot(x, y2, label='ROI')
plt.xlabel("Incentive")
plt.ylabel("ROI")
plt.title("ROI vs. Incentive")
```
