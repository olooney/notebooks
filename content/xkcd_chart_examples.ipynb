```python
@np.vectorize
def soc(x):
    return sin(x) if x > 5 else cos(x)

plt.plot(linspace(0,10), soc(linspace(0,10)))


plt.annotate(
        'WHERE WE PREDITED\nTHE MARKET WOULD GO',
        xy=(5.1, cos(5) ), arrowprops=hand_drawn_arrow, xytext=(3, 0.5))

plt.annotate(
        'WHERE THE MARKET\nACTUALLY WENT',
        xy=(5.2, sin(5) + .05 ), arrowprops=hand_drawn_arrow, xytext=(6, -0.5))
```

```python
N = 20
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * np.random.rand(N)
width = 2 * np.random.rand(N)

ax = plt.subplot(111) #, polar=True)
bars = ax.bar(theta, radii, width=width, bottom=0.0)

# Use custom colors and opacity
for r, bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.jet(r / 10.))
    bar.set_alpha(0.5)

plt.show()
```
